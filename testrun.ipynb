```python
import pandas as pd
import sys
import csv
import requests
import os
import sys
import time
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from fastai.vision.widgets import *
from fastai.vision.all import *
```

```python
learn_inf = load_learner("/content/trained.pkl")
```

```python
from ipywidgets import widgets

uploader = widgets.FileUpload()
display(uploader)

def on_upload_change(change):
    img = PILImage.create(uploader.data[0])

    result = learn_inf.predict(img)
    #result

    z = str(result[1])

    found_index = int(z[str(result[1]).index("(")+1:len(z)-1])

    found = str(result[0])

    percentage_str = str(result[2][found_index])

    percentage = float(percentage_str[percentage_str.index("(")+1:len(percentage_str)-1]) * 100

    print(found + " @ " + str(percentage) + "%")
    img.show()

uploader.observe(on_upload_change, names='_counter')
```
